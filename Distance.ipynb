```python
import bqplot as bqp
import ipywidgets as ipw
import numpy as np
import io
```

```python
FileUpload1 = ipw.FileUpload(
              accept = '.jpg',
              multiple = False,
              description = 'Subir imagen'
)
BoundedFloatText1 = ipw.BoundedFloatText(description = 'x1:', layout = ipw.Layout(width = '150px', padding = '2px 0px 0px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText2 = ipw.BoundedFloatText(description = 'y1:', layout = ipw.Layout(width = '150px', padding = '4px 0px 0px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText3 = ipw.BoundedFloatText(description = 'x2:', layout = ipw.Layout(width = '150px', padding = '4px 0px 2px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText4 = ipw.BoundedFloatText(description = 'y2:', layout = ipw.Layout(width = '150px', padding = '2px 0px 2px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText5 = ipw.BoundedFloatText(description = 'x3:', layout = ipw.Layout(width = '150px', padding = '2px 0px 0px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText6 = ipw.BoundedFloatText(description = 'y3:', layout = ipw.Layout(width = '150px', padding = '4px 0px 0px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText7 = ipw.BoundedFloatText(description = 'x4:', layout = ipw.Layout(width = '150px', padding = '4px 0px 2px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText8 = ipw.BoundedFloatText(description = 'y4:', layout = ipw.Layout(width = '150px', padding = '2px 0px 2px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText9 = ipw.BoundedFloatText(layout = ipw.Layout(width = '150px', padding = '2px 0px 2px 3px', margin = '0px 0px 0px 0px', border = '0px'))
BoundedFloatText10 = ipw.BoundedFloatText(layout = ipw.Layout(width = '150px', padding = '2px 0px 0px 3px', margin = '0px 0px 0px 0px', border = '0px'))
WLabel1 = ipw.Label(value = 'Escala:')
WLabel2 = ipw.Label(value = 'Distancia:')
WLabel3 = ipw.Label(value = 'Unidades:')
WLabel4 = ipw.Label(value = 'Desarrollado por:')
WLabel5 = ipw.Label(value = 'Junior Aguilar Hancco')
Text1 = ipw.Text(value = 'Unidades', layout = ipw.Layout(width = '150px', padding = '2px 0px 10px 3px', margin = '0px 0px 0px 0px', border = '0px'))

scale_x1  = bqp.LinearScale(min = 0, max = 1)
scale_y1  = bqp.LinearScale(min = 0, max = 1)

Lines1 = bqp.Lines(x = [0,0.5], y = [0,0.5],
         scales = {'x': scale_x1, 'y': scale_y1,},
         opacities = [1],
         colors = ['black'],
         stroke_width = 3,
         line_style = 'solid'
)
Lines2 = bqp.Lines(x = [], y = [],
         scales = {'x': scale_x1, 'y': scale_y1,},
         opacities = [1],
         colors = ['black'],
         stroke_width = 3,
         line_style = 'solid'
)
Lines3 = bqp.Lines(x = [], y = [],
         scales = {'x': scale_x1, 'y': scale_y1,},
         opacities = [1],
         colors = ['black'],
         stroke_width = 2,
         line_style = 'dashed'
)
Scatter1 = bqp.Scatter(x = [], y = [],
                   scales = {'x': scale_x1, 'y': scale_y1,},
                   opacities = [1],
                   colors = ['black'],
                   marker = 'circle',
                   default_size = 200
)
Scatter2 = bqp.Scatter(x = [], y = [],
                   scales = {'x': scale_x1, 'y': scale_y1,},
                   opacities = [1],
                   colors = ['black'],
                   marker = 'circle',
                   default_size = 200
)
Label1 = bqp.Label(x = [], y = [], text = [], default_size = 20, font_weight = 'bolder', colors = ['black'])
Label2 = bqp.Label(x = [], y = [], text = [], default_size = 20, font_weight = 'bolder', colors = ['black'])

ax_x1     = bqp.Axis(scale = scale_x1, label = '', color='white', grid_lines = 'none', visible = False)
ax_y1     = bqp.Axis(scale = scale_y1, label = '', color='white', orientation = 'vertical', grid_lines = 'none', visible = False)
scales1   = {"x": scale_x1, "y": scale_y1}
m_fig     = dict(left = 0, top = 0, bottom = 0, right = 0)
w         = str(1360 / 1.186) + 'px'
h         = str(767 / 1.186) + 'px'
fig1      = bqp.Figure(marks = [], axes = [ax_x1, ax_y1], fig_margin = m_fig, padding_x = 0, padding_y = 0, layout = ipw.Layout(width = w, height = h))

Button1   = ipw.Button(description = 'Mostrar imagen')
Button2   = ipw.Button(description = 'Dibujar escala')
Button3   = ipw.Button(description = 'Definir linea')
Button4   = ipw.Button(description = 'Medir')
Button5   = ipw.Button(description = 'Descargar imagen')
VBox1     = ipw.VBox([FileUpload1, Button1, Button2, BoundedFloatText1, BoundedFloatText2, BoundedFloatText3, BoundedFloatText4, Button3, 
                      BoundedFloatText5, BoundedFloatText6, BoundedFloatText7, BoundedFloatText8, Button4, WLabel1, BoundedFloatText9, 
                      WLabel2, BoundedFloatText10, WLabel3, Text1, Button5, WLabel4, WLabel5])
HBox1     = ipw.HBox([VBox1, fig1])
display(HBox1)

def SetImage(value):
    global imagen1
    im        = io.BytesIO(FileUpload1.data[0])
    raw_image = im.read()   
    ipyimage  = ipw.Image(value = raw_image, format = "jpg")
    imagen1   = bqp.Image(image = ipyimage, scales = scales1)
    imagen1.on_element_click(print_event)
    fig1.marks = [imagen1]
Button1.on_click(SetImage)
def print_event(self, target):
    global cords
    cords = target
click1 = 0
def Escala(value):
    global click1
    click1 = click1 + 1
    if click1 % 2 != 0:
        BoundedFloatText1.value = cords.get('data').get('click_x')
        BoundedFloatText2.value = cords.get('data').get('click_y') 
    else:
        BoundedFloatText3.value = cords.get('data').get('click_x')
        BoundedFloatText4.value = cords.get('data').get('click_y') 
    Lines1.x = np.array([BoundedFloatText1.value, BoundedFloatText3.value])
    Lines1.y = np.array([BoundedFloatText2.value, BoundedFloatText4.value])
    Scatter1.x = [BoundedFloatText1.value, BoundedFloatText3.value]
    Scatter1.y = [BoundedFloatText2.value, BoundedFloatText4.value]
    fig1.marks = [imagen1, Lines1, Lines2, Scatter1, Scatter2]
Button2.on_click(Escala)
click2 = 0
def Linea(value):
    global click2
    click2 = click2 + 1
    if click2 % 2 != 0:
        BoundedFloatText5.value = cords.get('data').get('click_x')
        BoundedFloatText6.value = cords.get('data').get('click_y') 
    else:
        BoundedFloatText7.value = cords.get('data').get('click_x')
        BoundedFloatText8.value = cords.get('data').get('click_y') 
    Lines2.x = np.array([BoundedFloatText5.value, BoundedFloatText7.value])
    Lines2.y = np.array([BoundedFloatText6.value, BoundedFloatText8.value])
    Scatter2.x = [BoundedFloatText5.value, BoundedFloatText7.value]
    Scatter2.y = [BoundedFloatText6.value, BoundedFloatText8.value]
    fig1.marks = [imagen1, Lines1, Lines2, Scatter1, Scatter2]
Button3.on_click(Linea)
xc = 0.1 * np.cos(np.linspace(0, 2 * np.pi, 50))
yc = 0.1 * np.sin(np.linspace(0, 2 * np.pi, 50)) * (1360/767)
def Medir(value):
    factor = BoundedFloatText9.value / abs(BoundedFloatText3.value - BoundedFloatText1.value)
    dist_x = abs(BoundedFloatText5.value - BoundedFloatText7.value) * factor
    dist_y = abs(BoundedFloatText6.value - BoundedFloatText8.value) / (1360/767) * factor
    BoundedFloatText10.value = np.sqrt(dist_x ** 2 + dist_y ** 2)
    Lines3.x = xc + cords.get('data').get('click_x')
    Lines3.y = yc + cords.get('data').get('click_y') 
    Label1.x = [(BoundedFloatText1.value + BoundedFloatText3.value) / 2]
    Label1.y = [(BoundedFloatText2.value + BoundedFloatText4.value) / 2]
    Label1.text = ['Linea de escala: ' + str(round(BoundedFloatText9.value, 2)) + ' ' + Text1.value]
    Label2.x = [(BoundedFloatText5.value + BoundedFloatText7.value) / 2]
    Label2.y = [(BoundedFloatText6.value + BoundedFloatText8.value) / 2]
    Label2.text = ['Distancia a red gasificada: ' + str(round(BoundedFloatText10.value, 2)) + ' ' + Text1.value]
    fig1.marks = [imagen1, Lines1, Lines2, Lines3, Scatter1, Scatter2, Label1, Label2]
Button4.on_click(Medir)
def Descargar(value):
    fig1.save_png("Distancia.png")
Button5.on_click(Descargar)
```
